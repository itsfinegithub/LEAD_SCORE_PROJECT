```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
dataset=pd.read_csv('/home/ubuntu/Downloads/educationcompany/Leads.csv')
```

```python
dataset.head(5)
```

```python
# finding the missing values
dataset.isna().sum().sort_values(ascending=False)
```

```python
# deleting the columns which is having more missing values
dataset.drop(columns=['Lead Quality','Asymmetrique Activity Index',
                      'Asymmetrique Profile Score','Asymmetrique Activity Score',
                      'Asymmetrique Profile Index','Prospect ID','Lead Number'],inplace=True)
```

```python
dataset.info()
```

```python
# finding the unique values in categorical columns
categ=dataset.select_dtypes('object').columns

for i in categ:
    print(dataset[i].value_counts())
    print()
```

```python
dataset=dataset.replace('google','Google')
dataset=dataset.replace('Want to take admission but has financial problems','Shall take in the next coming month')
dataset=dataset.replace('Shall take in the next coming month','Interested in Next batch')
dataset=dataset.replace('Still Thinking','In confusion whether part time or DLP')
```

```python
# dropping skewed columns
dataset.drop(columns=['What matters most to you in choosing a course','Search',
                      'Magazine','Newspaper Article','X Education Forums','Newspaper',
                      'Digital Advertisement','Through Recommendations','Receive More Updates About Our Courses',
                      'Update me on Supply Chain Content','Update me on Supply Chain Content',
                      'I agree to pay the amount through cheque','Do Not Call',
                      'What matters most to you in choosing a course','Get updates on DM Content'],inplace=True)
```

```python
# replacing the null values with NaN
dataset=dataset.replace(['unknown','Select'],np.NaN)
```

```python
categ=dataset.select_dtypes('object').columns

for i in categ:
    print(dataset[i].value_counts())
    print()
```

```python
categ=dataset.select_dtypes('object').columns
```

```python
# replacing  the missing values of categorical and numerical columns
for i in categ:
    dataset[i].fillna(dataset[i].mode()[0],inplace=True)
```

```python
dataset.isna().sum().sort_values(ascending=False)
```

```python
# numerical columns
numeric=dataset.select_dtypes('number')
numeric.isna().sum()
```

```python
# replacing the missing values of numerical columns
for j in numeric:
    dataset[j].fillna(dataset[j].median(),inplace=True)
```

```python
numeric.isna().sum()
```

```python
categorical_features = dataset.select_dtypes(include=[np.object])
```

```python
categorical_features.nunique()
```

```python
# finding and applying label encoder to the  columns which is having two unique values
cat = []
for i in categorical_features.columns:
    if categorical_features[i].nunique()==2:
        cat.append(i)
print(cat)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
for i in cat:
    dataset[i]= encoder.fit_transform(dataset[i])
```

```python
dataset.head(3)
```

```python
# dropping columns which are not important
dataset.drop(columns=['Country','City'],inplace=True)
```

```python
categorical_features = dataset.select_dtypes(include=[np.object]).columns
```

```python
#  applying the one-hot encoding for the columns which are having more than 2 unique values
dataset=pd.get_dummies(data=dataset,columns=categorical_features,drop_first=True)
```

```python
dataset
```

```python
# independent variables
x=dataset.drop(columns=['Converted'])
```

```python
x.head()
```

```python
# target variable
y=dataset['Converted']
```

```python

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=25)
```

```python
x_train.columns
```

```python
x_train
```

```python
x_train.mean()
```

```python
x_train.std()
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
x_train[['TotalVisits','Total Time Spent on Website','Page Views Per Visit']] = scaler.fit_transform(x_train[['TotalVisits','Total Time Spent on Website','Page Views Per Visit']])
x_test[['TotalVisits','Total Time Spent on Website','Page Views Per Visit']] = scaler.transform(x_test[['TotalVisits','Total Time Spent on Website','Page Views Per Visit']])
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier()
#  import rfe model for feature selection
from sklearn.feature_selection import RFE

rfe= RFE(rf_model,n_features_to_select=10)

rfe.fit(x_train,y_train)
```

```python
pd.DataFrame(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
top_10_col=x_train.columns[rfe.support_]
top_10_col
```

```python
x_train = x_train[top_10_col]
x_train
```

```python
x_test = x_test[top_10_col]
x_test
```

```python
x_train.columns=['TotalVisits', 'Total_Time_Spent_on_Website', 'Page_Views_Per_Visit',
       'Lead_Origin_Lead_Add_Form', 'Last_Activity_SMS_Sent',
       'What_is_your_current_occupation_Working_Professional',
       'Tags_Closed_by_Horizzon', 'Tags_Lost_to_EINS',
       'Tags_Will_revert_after_reading_the_email',
       'Last_Notable_Activity_SMS_Sent']
```

```python
x_test.columns=['TotalVisits', 'Total_Time_Spent_on_Website', 'Page_Views_Per_Visit',
       'Lead_Origin_Lead_Add_Form', 'Last_Activity_SMS_Sent',
       'What_is_your_current_occupation_Working_Professional',
       'Tags_Closed_by_Horizzon', 'Tags_Lost_to_EINS',
       'Tags_Will_revert_after_reading_the_email',
       'Last_Notable_Activity_SMS_Sent']
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train,y_train)
lr_pred = lr_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,plot_roc_curve
```

```python
print(classification_report(y_test,lr_pred))
```

```python
confusion_matrix(y_test,lr_pred)
```

```python
accuracy_score(y_test,lr_pred)
```
